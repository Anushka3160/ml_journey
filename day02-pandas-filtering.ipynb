```python
# making it more feasible. The head or tail func. along with the sort_values func. can be totally ignored by using nlargest and nsmallest(BUT ONLY ON NUMERIC VALUES)
df.nlargest(10, "age")
```

```python
df.nsmallest(10,["attendance_rate","study_hours_per_week"])
```

```python
# Students with:
# Age > 18
# Marks > 70
# Sorted by Marks descending
# Top 2 results
df[(df["age"]>18) & (df["final_score"]>70)].nlargest(2,"final_score")
```

```python
# Top 3 youngest females with score > 60
df[(df["final_score"]>60) & (df["gender"]=="Female")].nsmallest(3,"age")
```

```python
# Top 2 highest scoring females under age 25
df[(df["age"]<25) & (df["gender"]=="Female")].nlargest(2,"final_score")
```
