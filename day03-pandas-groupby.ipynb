```python
df=pd.read_csv("/kaggle/input/datasets/ziya07/college-student-health-behavior-dataset/student_health_dataset_50k.csv")
df
```

```python
df.groupby("sleep_quality")["heart_rate"].mean()
```

```python
df.groupby(["sleep_quality","diet_type"])["heart_rate"].mean()
```

```python
df.groupby(["sleep_quality","diet_type","gender"])["heart_rate"].mean()
```

```python
df.groupby(["sleep_quality","diet_type"])["heart_rate"].agg(["mean","max","median","count"])
```

```python
df.groupby(["sleep_quality","diet_type","gender"])["heart_rate"].agg(
    ag_mean="mean",
    maximum="max",
    middle="median",
    total="count")
```

```python
df.groupby(["gender","diet_type","health_condition"]).agg({
    "water_intake":["mean","max","min","count"],
    "heart_rate":["mean","max"],
    "sleep_duration":"mean",
})
```

```python
summary=df.groupby(["diet_type","sleep_quality"]).agg({
    "water_intake":["mean","max","min"],
    "heart_rate":["mean"],
    "sleep_duration":"mean"
})
largest=summary.nlargest(2,("water_intake","mean"))
largest
```
